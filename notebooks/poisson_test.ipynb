# Poisson model

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
p = np.append(np.ones(50,)*0.075, np.ones(50,)*0.25)
tmax = 100
rng = np.random.random(tmax, )
events = rng < p
```

```python
fig, axs = plt.subplots(nrows=2, ncols=1)
axs[0].bar(np.arange(tmax, ), events)
axs[0].set_title("Events")
axs[0].set_xlabel("time")
axs[1].bar(np.arange(tmax, ), p)
axs[1].set_title("Probability vector")
axs[1].set_xlabel("time")
fig.tight_layout()
```
